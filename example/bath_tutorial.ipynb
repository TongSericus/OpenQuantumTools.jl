# Tutorial on bath object
## Ohmic bath
Ohmic bath is the simplest build bath model in the package, you can define an Ohmic bath by

```julia
η = 1e-4; fc=4; T=16
bath = Ohmic(η, fc, T);
```

and plot out the spectrum and Lambshift by

```julia
p1 = plot(bath, :γ, range(0,20,length=200), label="", size=(800, 400), linewidth=2)
p2 = plot(bath, :S, range(0,20,length=200), label="", size=(800, 400), linewidth=2)
plot(p1, p2, layout=(1,2))
```

The correlation function can be calculated with a standard method `correlation`.

```julia
τ_axis = range(0,0.3,length=100)
c = [correlation(τ, bath) for τ in τ_axis]
plot(τ_axis, real.(c), linewidth=2, label="Real")
plot!(τ_axis, imag.(c), linewidth=2, label="Imag")
xlabel!(L"\tau \quad (\mathrm{ns})")
ylabel!(L"C(\tau)")
```
